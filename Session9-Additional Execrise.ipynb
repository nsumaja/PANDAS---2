# Filter out movies from the movies DataFrame that are of type 'Animation'.

```python
def filterGener(geners, gener_anmimation):
    res_lst = []
    for gen in geners:
        if(gen.find(gener_anmimation) != -1):
            res_lst.append(True)
        else:
            res_lst.append(False)
    return res_lst
        
Movies[filterGener(Movies.genres, 'Animation')]
```
